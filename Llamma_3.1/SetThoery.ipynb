```python
# Specify the model ID
model_id = "meta-llama/Meta-Llama-3.1-8B-Instruct"

# Initialize the text generation pipeline
text_gen_pipeline = pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto",
)
```

```python
def difficulty_generator(difficulty):
  if (difficulty == 0):
    question_instruction = """
      Below is an instruction that describes a task. Write a response that appropriately completes the request.

      ### Instruction:
      Generate an easy high school grade math problem.

      ### Response:
      """
  elif (difficulty == 2):
    question_instruction = """
      Below is an instruction that describes a task. Write a response that appropriately completes the request.

      ### Instruction:
      Generate a moderate high school grade math problem.

      ### Response:
      """
  elif (difficulty == 3):
    question_instruction = """
      Below is an instruction that describes a task. Write a response that appropriately completes the request.

      ### Instruction:
      Generate a challenging high school grade math problem.

      ### Response:
      """
  elif (difficulty == 4):
    question_instruction = """
      Below is an instruction that describes a task. Write a response that appropriately completes the request.

      ### Instruction:
      Generate a difficult high school grade math problem.

      ### Response:
      """
  elif (difficulty == 5):
    question_instruction = """
      Below is an instruction that describes a task. Write a response that appropriately completes the request.

      ### Instruction:
      Generate a extremely difficult high school grade math problem.

      ### Response:
      """
```

```python
# Prompt template for generating a math question
question_instruction = """
Below is an instruction that describes a task. Write a response that appropriately completes the request.

### Instruction:
Generate a challenging high school grade math problem.

### Response:
""",


# Generate the question
question_output = text_gen_pipeline(
    f"You are a helpful Maths assistant! Generate a math question.\n\n{question_instruction}",
    max_new_tokens=100,
)

# Extract the generated question text
question_text = question_output[0]["generated_text"]

# Display the generated question
display(Markdown(f"**Generated Question:**\n{question_text}"))
```

```python
# Prompt template for generating the answer to the generated question
answer_instruction = f"""
Below is a math problem. Solve the problem and provide a detailed explanation of the solution.

**Question:** {question_text}

### Response:
"""

# Generate the answer
answer_output = text_gen_pipeline(
    f"You are a helpful Maths assistant! Solve the problem and provide a detailed solution.\n\n{answer_instruction}",
    max_new_tokens=1028,
)

# Extract the generated answer text
answer_text = answer_output[0]["generated_text"]

# Display the generated answer
display(Markdown(f"**Generated Answer:**\n{answer_text}"))
```

```python
def difficulty_question_generator(difficulty, topic):
  if (difficulty == 0):
    question_instruction = f"""
      Below is an instruction that describes a task. Write a response that appropriately completes the request.

      ### Instruction:
      Generate an easy high school grade math problem.
      The topic is "{topic}" and you are supposed on generate the question on this topic
      Do not generate the answer.
      The question should only be mathematical, it should NOT contain any chemistry, physics or electrical topic based question.

      ### Response:
      """
  elif (difficulty == 2):
    question_instruction = f"""
      Below is an instruction that describes a task. Write a response that appropriately completes the request.

      ### Instruction:
      Generate a moderate high school grade math problem.
      The topic is "{topic}" and you are supposed on generate the question on this topic
      Do not generate the answer.
      The question should only be mathematical, it should NOT contain any chemistry, physics or electrical topic based question.

      ### Response:
      """
  elif (difficulty == 3):
    question_instruction = f"""
      Below is an instruction that describes a task. Write a response that appropriately completes the request.

      ### Instruction:
      Generate a challenging high school grade math problem.
      The topic is "{topic}" and you are supposed on generate the question on this topic
      Do not generate the answer.
      The question should only be mathematical, it should NOT contain any chemistry, physics or electrical topic based question.

      ### Response:
      """
  elif (difficulty == 4):
    question_instruction = f"""
      Below is an instruction that describes a task. Write a response that appropriately completes the request.

      ### Instruction:
      Generate a difficult high school grade math problem.
      The topic is "{topic}" and you are supposed on generate the question on this topic
      Do not generate the answer.
      The question should only be mathematical, it should NOT contain any chemistry, physics or electrical topic based question.

      ### Response:
      """
  elif (difficulty == 5):
    print('Ehe')
    question_instruction = f"""
      Below is an instruction that describes a task. Write a response that appropriately completes the request.

      ### Instruction:
      Generate a extremely difficult high school grade math problem based on "{topic}".
      It should be an extremely challenging Word Problem based on a fake scenario
      Do not generate the answer. Simply generate the question.
      The question should only be mathematical, it should NOT contain any chemistry, physics or electrical topic based question.

      ### Response:
      """
  return question_instruction
```

```python
def generate_math_problem(topic, difficulty):
    # Prompt template for generating a math question based on the topic and difficulty
    # question_instruction = f"""
    # Below is an instruction that describes a task. Write a response that appropriately completes the request.

    # ### Instruction:
    # Generate a high school level math problem on the topic "{topic}" with a difficulty level of {difficulty}.
    # Only generate the question, don't try to answer it
    # 1: "easy",
    # 2: "moderate",
    # 3: "challenging",
    # 4: "hard",
    # 5: "very hard"

    # Level 4 should be an easy word problem and level 5 should be a challenging word problem based on the {topic} with any fake scenario
    # It should be a purely mathematical problem only, there should not be any questions based on Chemistry, Electronics or Physics
    # ### Response:
    # """
    question_instruction = difficulty_question_generator(difficulty, topic)
    print(question_instruction)
    # Generate the question
    question_output = text_gen_pipeline(
        f"You are an experienced Math professor! Generate a math question.\n\n{question_instruction}",
        max_new_tokens=200,
    )

    # Extract the generated question text
    question_text = question_output[0]["generated_text"].split("### Response:")[-1].strip()

    # Display the generated question
    display(Markdown(f"**Generated Question:**\n{question_text}"))

    # Prompt template for generating the answer to the generated question
    answer_instruction = f"""
    Below is a math problem.
    Solve the problem and provide a detailed explanation of the solution with chain of thought

    **Question:** {question_text}

    ### Response:
    """

    # Generate the answer
    answer_output = text_gen_pipeline(
        f"You are an intelligent Math Professor. Solve the problem and provide a detailed solution.\n\n{answer_instruction}",
        max_new_tokens=1028,
    )

    # Extract the generated answer text
    answer_text = answer_output[0]["generated_text"].split("### Response:")[-1].strip()

    # Display the generated answer
    display(Markdown(f"**Generated Answer:**\n{answer_text}"))
```
